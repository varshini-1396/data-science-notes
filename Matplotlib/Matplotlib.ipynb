**Matplotlib example workflow**

```python
#0. Import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Set up plot
fig,ax = plt.subplots(figsize=(10,10))

#3. Plot data
ax.plot(x,y)
plt.show;
#Customize plot
ax.set(title="sample plot",
       xlabel="x-axis",
       ylabel="y-axis")
#save and show plot 
fig.savefig("sample-image.png")
plt.show();
```

```python
import numpy as np
```

```python
#Create some data 
x = np.linspace(0,10,100) #Returns 100 evenly spaced numbers over a specified interval.
x.size, x[:10]
```

***Line plot***

```python
#plot the data and create a line plot
fig,ax = plt.subplots()
ax.plot(x,x**2)
plt.show();
```

***Scatter plot***

```python
#Use the same data for scatter plot
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x));
plt.show()
```

```python
#Another scatter plot 
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
plt.show();
```

***Bar plot***

```python
#plotting from a dictionary
nut_butter_prices= {"Peanut Butter":10,
                    "Almond Butter":8,
                    "Cashew Butter":13}
fig,ax= plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Nut Butter store",
       xlabel= "Butter",
       ylabel="Prices")
plt.show();
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
plt.show();
```

***Histogram***

```python
x= np.random.randn(1000) #Return a sample (or samples) from the "standard normal" distribution
x.size, x[:100]
```

```python
fig,ax = plt.subplots()
ax.hist(x);
plt.show();
```

***Subplots -> can be done in two ways***

```python
#Subplots option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5));
ax1.plot(x,x/2);
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.scatter(np.random.randn(100),np.random.randn(100));
ax4.hist(x);
plt.show();
```

```python
#Subplots option 2
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5));
ax[0,0].plot(x,x/2);
ax[0,1].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,0].scatter(np.random.randn(100),np.random.randn(100));
ax[1,1].hist(x);
plt.show();
```

**Plotting from Pandas DataFrame**

```python
#generates a time series of 1000 random values indexed by dates starting from '1/1/2024'
#computes their cumulative sum using .cumsum(), and plots the resulting time series. plt.show() displays the plot on the screen.

ts= pd.Series(np.random.randn(1000),index=pd.date_range('1/1/2024',periods=1000))
ts=ts.cumsum()
ts.plot();
plt.show()
```

```python
over_50= heart_disease[heart_disease["age"]>50]
len(over_50)
```

```python
over_50.head()
```

```python
#Pyplot method
heart_disease.plot(x="age",y="chol",c="target",kind="scatter");
plt.show();
```

```python

fig,ax= plt.subplots(figsize=(10,6))
over_50.plot(x="age",
             y="chol",
             c="target",
             kind="scatter",
             ax=ax);
#ax.set_xlim(40,90) changing x limits
plt.show();
```

```python
# OO method from scratch
fig,ax= plt.subplots(figsize=(10,6))
ax.set(title="Heart disease and Cholestrol levels",
       xlabel="Age",
       ylabel="Cholestrol")
#Customize the plot
scatter=ax.scatter(x=over_50["age"],
              y=over_50["chol"],
              c=over_50["target"])
#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");
#Add a Horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--")
plt.show();
```

```python
heart_disease.head()
```

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.set(title="Just a plot",
       xlabel="Age",
       ylabel="thelach")
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_50["thalach"].mean(),linestyle="--")
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["thalach"],
                  c=over_50["target"])
plt.show();
```

```python
#Subplot of chol, age, thalach.
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,12),
                              sharex=True)
ax0.set(title="Heart Diease and Cholestrol",
        ylabel="Cholestrol")
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"]);
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(heart_disease["chol"].mean(),linestyle="--")

ax1.set(title="Heart Diease and thalach",
        xlabel="Age",
        ylabel="thalach")
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(heart_disease["thalach"].mean(),linestyle="--")
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold");
plt.show();                
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

# OO method from scratch
fig,ax= plt.subplots(figsize=(10,6))
ax.set(title="Heart disease and Cholestrol levels",
       xlabel="Age",
       ylabel="Cholestrol")
#Customize the plot
scatter=ax.scatter(x=over_50["age"],
              y=over_50["chol"],
              c=over_50["target"],
               cmap="winter") #cmap chanegs the colour scheme
#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");
#Add a Horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--")
plt.show();
```

```python
#Subplot of chol, age, thalach.
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,12),
                              sharex=True)
ax0.set(title="Heart Diease and Cholestrol",
        ylabel="Cholestrol")
#Changing x axis limits
ax0.set_xlim([50,80])
ax1.set_xlim([50,80])
#Changing y axis limits
ax0.set_ylim([60,200])
ax1.set_ylim([60,200])
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                   cmap="winter");
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(heart_disease["chol"].mean(),linestyle="--")

ax1.set(title="Heart Diease and thalach",
        xlabel="Age",
        ylabel="thalach")
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                   cmap="winter")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(heart_disease["thalach"].mean(),linestyle="--")
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold");
plt.show();                
```
